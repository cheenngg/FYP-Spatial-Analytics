# Simple code to convert datasets in .mat extenstion to a dataframe 

```python
import scipy.io
import pandas as pd

# Load your .mat file
mat = scipy.io.loadmat('HER-Master/datasets/LR0_input_randomfield_cal1000.mat')

# Assuming the .mat file contains a variable 'data' that you want to convert
# You might need to adjust 'data' to the actual variable name you're interested in
df = pd.DataFrame(mat['data'])
display(df)
```
