```python
def f_plot_prediction(z_plot, z_entropy_plot, pmf_pred, x_target, y_target, x_target_grid, y_target_grid, x, y, z, idx_cal, her, x_pmf, y_pmf, true_on, shp_basin=None):
    randplot_set = []
    for i in range(len(x_pmf)):
        randplot_set.append(np.where(np.logical_and(np.isin(x_target, x_pmf[i]), np.isin(y_target, y_pmf[i])))[0][0])
    ncols = 2
    nrows = int(np.ceil((len(x_pmf) / ncols)))
    fig, axs = plt.subplots(nrows, ncols, figsize=(15, 15))
    fig.tight_layout(pad=5.0)
    for i, ax in enumerate(axs.flat):
        if i < len(x_pmf):
            ax.pcolor(x_target_grid, y_target_grid, z)
            ax.scatter(x[idx_cal], y[idx_cal], 1+70*(z[idx_cal]-np.min(z))/(np.max(z)-np.min(z)), 'k+')
            ax.scatter(x_pmf, y_pmf, 100, 's', markeredgecolor=[0.9, 0.9, 0.9], linewidth=1.5)
            ax.set_title('Original\n\n')
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            ax.set_aspect('equal', adjustable='box')
            ax.set_xlim([np.min(x_target), np.max(x_target)])
            ax.set_ylim([np.min(y_target), np.max(y_target)])
            ax.colorbar()
            if shp_basin is not None:
                ax.add_shapefile(shp_basin, facecolor='none', edgecolor=[0.9, 0.9, 0.9], linewidth=2)
        elif i < 2*len(x_pmf):
            ax.pcolor(x_target_grid, y_target_grid, z_plot)
            ax.scatter(x[idx_cal], y[idx_cal], 1+70*(z[idx_cal]-np.min(z))/(np.max(z)-np.min(z)), 'k+')
            ax.scatter(x_pmf, y_pmf, 100, 's', markeredgecolor=[0.9, 0.9, 0.9], linewidth=1.5)
            ax.set_title('HER E-type\n\n')
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            ax.set_aspect('equal', adjustable='box')
            ax.set_xlim([np.min(x_target), np.max(x_target)])
            ax.set_ylim([np.min(y_target), np.max(y_target)])
            ax.colorbar()
            if shp_basin is not None:
                ax.add_shapefile(shp_basin, facecolor='none', edgecolor=[0.9, 0.9, 0.9], linewidth=2)
        else:
            ax.pcolor(x_target_grid, y_target_grid, z_entropy_plot)
            ax.scatter(x[idx_cal], y[idx_cal], 1+70*(z[idx_cal]-np.min(z))/(np.max(z)-np.min(z)), 'k+')
            ax.scatter(x_pmf, y_pmf, 100, 's', markeredgecolor=[0.9, 0.9, 0.9], linewidth=1.5)
            ax.set_title('Entropy map [bits]\n\n')
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            ax.set_aspect('equal', adjustable='box')
            ax.set_xlim([np.min(x_target), np.max(x_target)])
            ax.set_ylim([np.min(y_target), np.max(y_target)])
            ax.colorbar()
            if shp_basin is not None:
                ax.add_shapefile(shp_basin, facecolor='none', edgecolor=[0.9, 0.9, 0.9], linewidth=2)
    plt.suptitle('Maps | Dataset: ' + her['txt'])
    plt.legend(['H', 'cal', 'selected PMF'], loc='northwest')
    plt.linkaxes(axs, 'xy')
    plt.show()
    if true_on == 'N':
        for target in randplot_set:
            plt.figure()
            plt.bar(her['bin_centers_edges_z'], pmf_pred[target], 'k')
            plt.xlim([-10, 10])
            plt.ylim([0, 0.2])
            plt.xlabel('z')
            plt.title('target ' + str(target) + '\nx=' + str(x_target[target]) + '; y=' + str(y_target[target]) + '\nH = ' + str(round(f_entropy(pmf_pred[target]), 2)) + ' bits')
            plt.gca().set_aspect('equal', adjustable='box')
    elif true_on == 'Y':
        for target in randplot_set:
            plt.figure()
            plt.bar(her['bin_centers_edges_z'], pmf_pred[target], 'k')
            plt.hold(True)
            plt.plot([z[target], z[target]], [0, 0.5], 'm--', linewidth=1.5)
            plt.xlim([-10, 10])
            plt.ylim([0, 0.2])
            plt.xlabel('z')
            plt.title('target ' + str(target) + '\nx=' + str(x_target[target]) + '; y=' + str(y_target[target]) + '\nH = ' + str(round(f_entropy(pmf_pred[target]), 2)) + ' bits')
            plt.gca().set_aspect('equal', adjustable='box')
    plt.show()


```
