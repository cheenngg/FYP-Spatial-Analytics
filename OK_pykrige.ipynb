```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pykrige.kriging_tools as kt
from pykrige.ok import OrdinaryKriging
```

```python
crimes_df = pd.read_csv("crimes_processed.csv")
meuse_df = pd.read_csv("meuse_processed.csv")
```

```python
meuse_df
```

```python
meuse_df.describe()
```

```python
x = np.array(meuse_df['x'])
y = np.array(meuse_df['y'])
zinc = np.array(meuse_df['zinc'])
```

```python
cax = plt.scatter(x, y, c=zinc)
cbar = plt.colorbar(cax, fraction=0.03)
plt.title('Measured Zinc Levels')
```

```python
OK = OrdinaryKriging(
    x, 
    y, 
    zinc, 
    variogram_model='exponential',
    verbose=True,
    enable_plotting=True,
    nlags=10,
)
```

```python
OK.variogram_model_parameters
```

```python
gridx = np.arange(178605, 181390, 10, dtype='float64')
gridy = np.arange(329714, 333611, 10, dtype='float64')
zstar, ss = OK.execute("grid", gridx, gridy)
```

```python
# Create a figure with a specific size
fig = plt.figure(figsize=(12, 10))

# Plot the data
cax = plt.imshow(zstar, extent=(178605, 181450, 329714, 333611), origin='lower')
plt.scatter(x, y, c='k', marker='.')

# Set the limits of the x-axis to increase its length
plt.xlim(178700, 181450)

# Add colorbar
cbar = plt.colorbar(cax)

# Add labels and title
plt.title('Zinc estimate')

# Show the plot
plt.show()
```

```python
# Create a figure with a specific size
fig = plt.figure(figsize=(12, 10))

# Plot the data
cax = plt.imshow(np.sqrt(ss), extent=(178605, 181450, 329714, 333611), origin='lower', vmin = 0)
plt.scatter(x, y, c='k', marker='.')

# Set the limits of the x-axis to increase its length
plt.xlim(178700, 181450)

# Add colorbar
cbar = plt.colorbar(cax)

# Add labels and title
plt.title('Zinc standard devation')

# Show the plot
plt.show()
```
